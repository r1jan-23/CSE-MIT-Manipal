```python
#QN1 

# Creating movie data
movies_data = {
    "Movie Name": ["Dangal", "Inception", "Interstellar", "Zindagi Na Milegi Dobara", "Parasite",
                   "Avengers: Endgame", "3 Idiots", "Titanic", "Sholay", "The Dark Knight"],
    "Language": ["Hindi", "English", "English", "Hindi", "Korean", "English", "Hindi", "English", "Hindi", "English"],
    "Genre": ["Drama", "Sci-Fi", "Sci-Fi", "Comedy", "Thriller", "Action", "Comedy", "Romance", "Action", "Action"],
    "Rating": [8.4, 8.8, 8.6, 8.2, 8.6, 8.4, 8.4, 7.8, 8.2, 9.0],
    "Review": ["Inspiring", "Mind-bending", "Thought-provoking", "Feel-good", "Masterpiece", 
               "Epic", "Entertaining", "Emotional", "Classic", "Legendary"]
}
```

```python
# Reading the CSV file into a DataFrame
df = pd.read_csv("Movies.csv")
```

```python
# Finding the movie with the highest rating
highest_rated_movie = df.loc[df["Rating"].idxmax()]
print("\nMovie with the highest rating:")
print(highest_rated_movie)
```

```python
# Extracting all Hindi movies and saving them to 'HindiMovies.csv'
hindi_movies = df[df["Language"] == "Hindi"]
hindi_movies.to_csv("HindiMovies.csv", index=False)
print("\nHindiMovies.csv file created successfully.")
```

```python
#QN2
#a.Create a table with the 5 number summary of all the numeric attributes. 
import pandas as pd
import numpy as np

path_to_cereals = "Cereals.csv"
cereal_data = pd.read_csv(path_to_cereals)
cereal_data
```

```python

#b. For each of the numeric attributes identify and replace all missing data with the arithmetic mean of the attribute. 
cereal_data.describe().loc[['min', '25%', '50%', '75%', 'max']]
```

```python
#Replacing the missing data with the mean of the attribute
cereal_data = cereal_data.replace('?', np.nan)
cereal_data
```

```python
#c. Create a table with the 5 number summary of all the numeric attributes after treating missing values.
#d.For each numeric attribute,identify and replace all noisy data with the median of attribute.  
#e. Create a table with the 5 number summary of all the numeric attributes after treating noisy values.
l = ['carbo', 'sugars', 'potass']
cereal_data[l] = cereal_data[l].astype('float64')
print(cereal_data.dtypes)
cereal_data.describe().loc[['min', '25%', '50%', '75%', 'max']]
```

```python
numeric_columns = ['protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins']
cereal_data[numeric_columns] = cereal_data[numeric_columns].apply(lambda x: x.fillna(x.mean()))
cereal_data
```
