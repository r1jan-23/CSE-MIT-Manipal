```python
#Write a Python function to input two numbers and perform the Calculator operations of (+, -, *, /).

def no(var_a , var_b):          #Define the function 'no' to take two arguement
    var_c=var_a+var_b           # addition
    var_d=var_a-var_b           # substraction
    var_e=var_a*var_b           # multiplication
    var_f=var_a/var_b           # division
    return( var_c, var_d, var_e, var_f)     # Return the results of all operations as a tuple
```

```python
var_a=int(input("Enter 1st number: "))  
var_b=int(input("Enter 2nd Number: "))
print(no(var_a,var_b))                  
```

```python
#Write a Python function that takes an integer and returns True if itâ€™s a prime number and False otherwise.

def prime(var_x):
    flag=1
    for i in range(2,var_x):   # Check divisibility from 2 to var_x-1
        if(var_x%i==0):        # If var_x is divisible it is not prime
            flag=0
            break              # No need to check further
    return flag              
```

```python
var_x=int(input("Enter Number: "))
if(prime(var_x)==0):  # Check if the number is prime or not
    print("False")
else:
    print("True")
```

```python
# Create a Python function that creates a sequence between 1 and 100 and prints all the odd numbers. Compute and display the
# sum of all the even numbers. 

def odd():
    sum=0
    for i in range(1,100):   # Loop through numbers from 1 to 99
        if(i % 2 == 0):      # Check if the number is even 
            sum += i         #Add Even Number
        else:
            print(i)         #Print all odd number
    return sum
```

```python
a=odd()
print ("Sum of Even Number : ",a)  #Print Sum of even Nuber
```

```python
#Write a Python function that takes a string input from the user and counts the number of vowels and consonants in the string. 

def count_vowel(str_a):
    count_vowels=0
    count_consonants=0
    for i in str_a:   # Loop through each character in the string
        if i in "AEIOUaeiou":    # Check if the character is a vowel Uppercase and lowercase
            count_vowels+=1
        else:
            count_consonants+=1
    return(count_vowels,count_consonants)      # Return the counts of vowels and consonants as a tuple
```

```python
r=count_vowel(str(input("Enter String: ")) )
print("vowel is ",r[0]," & Consonants is ",r[1])
```
