```python
#import the libraries such as numpy, pandas, matplotlib and seaborn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Upload the dataset
data=pd.read_csv('deliveries.csv')
data
```

```python
#Q1
sns.barplot(x='batting_team', y='total_runs', data=data.groupby('batting_team', as_index=False)['total_runs'].sum())
plt.xlabel('Batting team')
plt.ylabel('Total runs')
plt.xticks(rotation=90)
plt.show()
```

```python
#Q2_a
#Finding top 10 batsman and their runs
print("Batsman & Their Runs")
print(data.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10))
```

```python
#Q2_b
#Converting the list of top10 batsman in visualization / pictorial form
sns.barplot(x=data.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10).index,y=data.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10).values)
plt.title('Top 10 batters')
plt.xlabel('batter')
plt.ylabel('total runs')
plt.xticks(rotation=90)
plt.show()
```

```python
#Q3_a
#Finding Top10 bowler name and their wickets
print(data.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(10))
```

```python
#Q3_b
#Coverting Top 10 bowlers name in visualization / pictorial form
sns.barplot(x=data.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(10).index,y=data.groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(10).values)
plt.title('Top 10 bowlers')
plt.xlabel("Bowler's Name ")
plt.ylabel("wicket's")
plt.xticks(rotation=90)
plt.show()
```

```python
#Q4_a
#Finding Team who #Q4_b
sns.barplot(x=data.groupby('batting_team')['extra_runs'].sum().index,y=data.groupby('batting_team')['extra_runs'].sum().values)
plt.title('total number of extras conceded by each team')
plt.xlabel('team')
plt.ylabel('runs')
plt.xticks(rotation=90)
plt.show()
print(data.groupby('batting_team')['extra_runs'].sum())
```

```python
#Q4_b
sns.barplot(x=data.groupby('batting_team')['extra_runs'].sum().index,y=data.groupby('batting_team')['extra_runs'].sum().values)
plt.title('total number of extras conceded by each team')
plt.xlabel('team')
plt.ylabel('runs')
plt.xticks(rotation=90)
plt.show()
```

```python
#Q5
run_outs = data[data['dismissal_kind']=='run out']
run_out_batsmen = run_outs['batter'].value_counts().head(5)
run_out_batsmen.plot(kind='bar', figsize=(12,6))
plt.title('top 5 batsmen involved in most run_outs')
plt.ylabel('Number of Run-outs')
plt.xlabel('Batsman')
plt.xticks(rotation=90)
plt.show()
```

```python
#Q6
extras_by_bowler = data.groupby('bowler')['extra_runs'].sum().sort_values(ascending=False).head(5)
extras_by_bowler.plot(kind='bar', figsize=(12,6))
plt.title('top 5 bowlers who conceded the most extras')
plt.ylabel('total extras')
plt.xlabel('Bowlers')
plt.xticks(rotation=90)
plt.show()
```

```python
     #Q7
dismissal_types = data['dismissal_kind'].value_counts()
dismissal_types.plot(kind='bar', figsize=(12,6))
plt.title('most common dismissal types')
plt.ylabel('frequency')
plt.xlabel('types')
plt.xticks(rotation=90)
plt.show()
```

```python
#Q8
over_runs = data.groupby('over')['total_runs'].sum()
over_balls = data.groupby('over')['ball'].count()
run_rate = over_runs / (over_balls / 6)
top_3_overs = run_rate.sort_values(ascending=False).head(3)
top_3_overs.plot(kind='bar', figsize=(12,6))
plt.title('top 3 overs has the highest run rate')
plt.ylabel('Run Rate')
plt.xlabel('top 3 Overs')
plt.xticks(rotation=0)
plt.show()
```

```python
#Q9
powerplay_data = data[data['over'] <= 6]
runs_powerplay = powerplay_data.groupby('inning')['total_runs'].sum()
runs_powerplay.plot(kind='bar', figsize=(12,6))
plt.title('powerplay performance (first 6 overs)')
plt.ylabel('Total Runs')
plt.xlabel('inning')
plt.xticks(rotation=0)
plt.show()
```

```python
#Q10
fours_sixes = data[(data['batsman_runs'] == 4) | (data['batsman_runs'] == 6)]
fours_sixes_count = fours_sixes.groupby('batter')['batsman_runs'].value_counts()
top_batsmen = fours_sixes_count.groupby('batter').sum().sort_values(ascending=False).head(5)
top_batsmen.plot(kind='bar', figsize=(12,6))
plt.title('Top 5 Batsmen with Most Fours & Sixes')
plt.ylabel('Total Fours & Sixes')
plt.xlabel('Batsmen')
plt.xticks(rotation=90)
plt.show()
```
