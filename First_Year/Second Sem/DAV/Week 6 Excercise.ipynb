```python
#Create a DataFrame with the following data and labels
dat_mtcars=pd.read_csv("mtcars.csv",index_col=0)
dat_mtcars
```

```python

#Q1.Explore the distribution of Fuel efficiency of the cars
# Remember to label both axis and put a title to the plot
dat_mtcars["mpg"].plot(kind='hist', bins=10, title="Histogram of the Distribution of MPG")
```

```python
#Explore the distribution of Fuel efficiency of the cars
dat_mtcars["mpg"].plot(kind='box')
```

```python
#Q4.Find the car with the worst horsepower  
min_hp=dat_mtcars["hp"].min()
dat_mtcars[dat_mtcars["hp"]==min_hp]
```

```python
#Q7.Which is the car with the best qsec?
max_qsec=dat_mtcars["qsec"].max()
dat_mtcars[dat_mtcars["qsec"]==max_qsec]
```

```python
#Q8. What is average MPG for manual vs. automatic cars? 
dat_man= dat_mtcars[dat_mtcars["am"]==1]["mpg"] # Manual cars are marked as 1
print("Average MPG of manual cars:",dat_man.mean())
dat_auto= dat_mtcars[dat_mtcars["am"]==0]["mpg"]
print("Average MPG of automatic cars:",dat_auto.mean())
```

```python
#Q9. Draw Side by Side box plot to understand the difference in fuel efficience of Manual vs Automatic cars.  Analyze and write about fuel efficiency in each group (manual vs. automatic).  
dat_mtcars.boxplot(column='mpg', by='am')
```

```python
#Q10. What is the relationship between the weight of the car and MPG? 
# Create scatter plot
# Remember to label both axis and put a title to the plot, write the units too
dat_x= dat_mtcars[["mpg","wt"]]
dat_x= dat_x.reset_index(drop=True)
dat_x.plot(kind='scatter', x='mpg', y='wt', color='blue', title='Scatter Plot from DataFrame')
```

```python
#Q13. Explain the relationship between horse power and number of cylinders in the car.
dat_x= dat_mtcars[["hp","cyl"]]
dat_x= dat_x.reset_index(drop=True)
dat_x.plot(kind='scatter', x='hp', y='cyl', color='blue', title='Scatter Plot from DataFrame')
    
```

```python
#Q15. Which car would I pick if I am looking for high speed as well as good fuel efficiency?
dat_mtcars["speed"]=dat_mtcars["qsec"]*dat_mtcars["mpg"]
max_speed=dat_mtcars["speed"].max()
dat_mtcars[dat_mtcars["speed"]==max_speed]
```

```python
#QN-2
#Load the dataset
dat_cereals=pd.read_csv("Cleaned_Cereals.csv")
dat_cereals
```

```python
#data summary
dat_cereals.describe()
```

```python
#Q1 How many cereals(unique brands) are there in the dataset?
dat_cereals["name"].nunique()
```

```python
#Q2 How many cereals are listed per manufacturer of the cereal?
dat_cereals["mfr"].value_counts()
```

```python
#Q3 How many hot vs cold cereals?
dat_cereals["type"].value_counts()
```

```python
#Q4 Which is the best and worst cereal?
max_rating=dat_cereals["rating"].max()
dat_cereals[dat_cereals["rating"]==max_rating]
```

```python
min_rating=dat_cereals["rating"].min()
dat_cereals[dat_cereals["rating"]==min_rating]
```

```python
#Q5 Compare the rating for hot vs cold cereals. Which do people prefer?
dat_hot= dat_cereals[dat_cereals["type"]=="H"]["rating"]
print("Average rating of hot cereals:",dat_hot.mean())
dat_cold= dat_cereals[dat_cereals["type"]=="C"]["rating"]
print("Average rating of cold cereals:",dat_cold.mean())
```

```python
#Q6 Which cereals have the highest fiber and lowest sugar?
max_fiber=dat_cereals["fiber"].max()
dat_cereals[dat_cereals["fiber"]==max_fiber]
```

```python
min_sugar=dat_cereals["sugars"].min()
dat_cereals[dat_cereals["sugars"]==min_sugar]
```

```python
#Q7 Which Cereals that have more than 3 grams of protein?
dat_cereals[dat_cereals["protein"]>3]
```

```python
#Q8 Tabulate and visualize Cereals by display Shelf.
dat_cereals["shelf"].value_counts()
```

```python
dat_cereals["shelf"].value_counts().plot(kind='bar')
```

```python
#Q9 How does the sugar content vary across different brands?
dat_cereals.groupby("mfr")["sugars"].mean()
```

```python
#create a box plot 
dat_cereals.boxplot(column='sugars', by='mfr')
```

```python
#Q10 What is the average calories in cereals per manufacturer?
dat_cereals.groupby("mfr")["calories"].mean()
```

```python
#Q11 What is the average nutritional content (calories, sugars, protein, fiber, etc.) across all cereals?
dat_cereals.iloc[:, 3:13].mean()
```

```python
#Q12 What is the relationship between sugar and calories?
dat_x= dat_cereals[["sugars","calories"]]  
dat_x= dat_x.reset_index(drop=True)
dat_x.plot(kind='scatter', x='sugars', y='calories', color='blue', title='Relationship between Sugar and Calories')
```

```python
#Q13 How does the sugar content compare in cereals with high vs. low ratings?
dat_cereals["rating"].mean()
dat_high= dat_cereals[dat_cereals["rating"]>dat_cereals["rating"].mean()]["sugars"]
print("Average sugar content in cereals with high ratings:",dat_high.mean())
dat_low= dat_cereals[dat_cereals["rating"]<dat_cereals["rating"].mean()]["sugars"]
print("Average sugar content in cereals with low ratings:",dat_low.mean())
```

```python
#Q14 Do cereals marketed as "healthy" (e.g., high fiber, low sugar) tend to have higher ratings?
dat_healthy= dat_cereals[(dat_cereals["fiber"]>dat_cereals["fiber"].mean()) & (dat_cereals["sugars"]<dat_cereals["sugars"].mean())]["rating"]
print("Average rating of healthy cereals:",dat_healthy.mean())
dat_unhealthy= dat_cereals[(dat_cereals["fiber"]<dat_cereals["fiber"].mean()) & (dat_cereals["sugars"]>dat_cereals["sugars"].mean())]["rating"]
print("Average rating of unhealthy cereals:",dat_unhealthy.mean())
```

```python
#Q15 Is there a relationship between rating and display shelf?
dat_x= dat_cereals[["rating","shelf"]]
dat_x= dat_x.reset_index(drop=True)
dat_x.plot(kind='scatter', x='rating', y='shelf', color='blue', title='Relationship between Rating and Display Shelf')
#There is no relationship between rating and display shelf.
#The ratings are spread across all the shelves.
#The ratings are not dependent on the display shelf.
```
