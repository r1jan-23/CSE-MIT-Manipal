```python
#QN-1

# Enter the size of the list 1D array
n = int(input("Enter the size of the list : "))

# Empty list to store the numbers
list_a = []

# Loop starting from 0 t0 (n-1)
for i in range(0, n):
    # Append each input number to the list
    list_a.append(int(input()))

# Display minimum value
print("Minimum Value is : ", min(list_a))

# Display maximum value 
print("Maximum Value is : ", max(list_a))

# Display sum 
print("Sum is", sum(list_a))

# Display average 
print("Average is", sum(list_a) / n)

#Enter a number for which the frequency is required
count1 = int(input("Enter the value of the number for which you require the frequency:"))

# Display the count (frequency) of the number in the list
print("Count of", count1, "is", list_a.count(count1))
```

```python
#QN-3

# Dictionary containing the year as key and the winning team as value
dict_wc = {
    1975: "West Indies", 1979: "West Indies", 1983: "India", 1987: "Australia", 
    1992: "Pakistan", 1996: "Sri Lanka", 1999: "Australia", 2003: "Australia", 
    2007: "Australia", 2011: "India", 2015: "Australia", 2019: "England", 
    2023: "Australia"
}

# Convert the dictionary values (team names) into list
list_wc = list(dict_wc.values())

# Create a set of unique teams from the list to remove duplicate
list_occur = set(list_wc)

# Initialize variables to find team of max wins
max_value=0
team = list_wc[0]

# Loop to count occurance in list
for i in list_occur:
    # If team has more occurrences than current max, update team and max_value
    if list_wc.count(i) > max_value:
        max_value = list_wc.count(i)
        team = i

# Print the most successful cricket team base on highest number of wins
print("Most Successful cricket team is", team)
```

```python
#QN-4

# input a sentence
sentence = input("Enter a Sentence: ")

# Convert the sentence to lowercase and split it to a list of words
list_word = list(sentence.lower().split(" "))

# Convert the list of words into a set to get unique words remove duplicate
set_word = set(list_word)

# Initialize an empty dictionary to store word counts
dict_word = {}

# Loop through each unique word in the set
for word in set_word:
    # Count the occurrences of each word in the list and store in the dictionary
    dict_word[word] = list_word.count(word)

# Print the dictionary with word counts
print(dict_word)
```

```python
#QN-5

# Initialize two empty lists to hold the input values for two sets
list_x = []
list_y = []

# Enter the size of the first set
list_x_len = int(input("Enter the size of the first set : "))
print("Input First Set ")

# Input values for the second set range (0 to list_x_len - 1)
for i in range(0, list_x_len):
    list_x.append(int(input()))  # Append each element to the list_x

# Enter the size of the second set
list_y_len = int(input("Enter the size of the second set : "))
print("Input Second Set")

# Input values for the second set range (0 to list_y_len - 1)
for i in range(0, list_y_len):
    list_y.append(int(input()))  # Append each element to the list_y

# Convert both lists into sets
set_x = set(list_x)
set_y = set(list_y)

# Print the first set
print("Set 1 is ", set_x)
print("Set 2 is ", set_y)

# Perform union operation (set_x ∪ set_y)
print("Union of Set 1 and Set 2 is:", set_x.union(set_y))

# Perform intersection operation (set_x ∩ set_y)
print("Intersection of Set 1 and Set 2 is:", set_x.intersection(set_y))

# Perform difference operation (set_x - set_y)
print("Difference of Set 1 - Set 2 is:", set_x.difference(set_y))

# Perform difference operation (set_y - set_x)
print("Difference of Set 2 - Set 1 is:", set_y.difference(set_x))
```
