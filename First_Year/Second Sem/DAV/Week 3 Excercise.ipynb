```python
#QN-1
import numpy as np

# Random integers between 1 and 50, with size of array being 3 rows and 4 columns
arr_1 = np.random.randint(1, 51, size=(3, 4))
print(arr_1)
```

```python
print("The mean =",np.mean(arr_1))                    # calculating mean

print("The median =",np.median(arr_1))                # calculating median

print("The Standard Deviation =",np.std(arr_1))       # calculating SD

print("The sum =",np.sum(arr_1))                      # calculating sum

print("Sum of each row:", np.sum(arr_1, axis=1))      # calculating sum of each row

print("The reshape matrix is :\n",arr_1.reshape(2,6))                             # Reshaping 3*4 matrix to 2*6
```

```python
import numpy as np

# Create two matrices of random values of size 3*3
arr_2 = np.random.randint(1, 10, size=(3, 3))
arr_3 = np.random.randint(1, 10, size=(3, 3))

# Print both matrix, addition, substraction, multiplication, dot product, transpose
print("Matrix A:\n", arr_2)
print("Matrix B:\n", arr_3)
print("\nMatrix Addition (A + B):\n", arr_2 + arr_3)
print("\nMatrix Subtraction (A - B):\n", arr_2 - arr_3)
print("\nElement-wise Multiplication (A * B):\n", arr_2 * arr_3)
print("\nDot Product (A . B):\n", np.dot(arr_2, arr_3))
print("\nTranspose of Matrix A:\n", np.transpose(arr_2))
print("\nTranspose of Matrix B:\n", np.transpose(arr_3))

# Determinant and Inverse of Matrix A
det_arr_2 = np.linalg.det(arr_2)
print("Determinant of Matrix A:", det_arr_2)
if not np.isclose(det_arr_2, 0):  # Check if determinant is not zero
    inv_arr_2 = np.linalg.inv(arr_2)
    print("Inverse of Matrix A:\n", inv_arr_2)
else:
    print("Determinant is 0, so inverse does not exist.")

# Determinant and Inverse of Matrix B
det_arr_3 = np.linalg.det(arr_3)
print("Determinant of Matrix B:", det_arr_3)
if not np.isclose(det_arr_3, 0):  # Check if determinant is not zero
    inv_arr_3 = np.linalg.inv(arr_3)
    print("Inverse of Matrix B:\n", inv_arr_3)
else:
    print("Determinant is 0, so inverse does not exist.")
```
